Случайная величина распределена равномерно на отрезке $[0, \theta]$. По выборке объема $n$ найдены оценки параметра $\theta$:

Случайная величина: $\xi \sim R(0, \theta)$

Выборка: $\overline{X}_{n} = (x_{1}, ..., x_{n})$

$\widetilde{\theta}_{1} = 2 \overline{x}$, 
$\widetilde{\theta}_{2} = x_{min}$, 
$\widetilde{\theta}_{3} = x_{max}$, 
$\widetilde{\theta}_{4} = x_{min} + x_{max}$, 
$\widetilde{\theta}_{5} = x_{1} + \frac{1}{n-1} \sum\limits_{k=2}^n x_{k}$.

а) Проверить оценки на несмещенность и состоятельность. Исправить эти оценки, если необходимо. Какая из исправленных оценок более эффективна?

б) Сгенерируйте выборку объема $N = 100$ для некоторого значения параметра $\theta$. Для всех значений $1 \le n \le N$ вычислите исправленные оценки и постройте графики $|\widetilde{\theta}_{k} - \theta|$ в зависимости от $n$.

```python
import numpy as np
from matplotlib import pyplot as plt
```

```python
theta = 150
selection_size = 100

X = np.random.randint(0, theta, size=selection_size)

theta_estimation = [
    lambda X: 2 * np.mean(X),
    lambda X: (len(X) + 1) * np.min(X),
    lambda X: (len(X) + 1) / len(X) * np.max(X),
    lambda X: np.min(X) + np.max(X),
    lambda X: X[0] + 1 / (len(X) + 1) * np.sum(X[1:])
]
```

```python
def theta_difference(theta_estimate, theta, X):
    diff = np.zeros(len(X))
    for i in range(0, len(X)):
        diff[i] = np.abs(theta_estimate(X[:i+1]) - theta)
    return diff
```

```python
fig, ax = plt.subplots()

for i in range(len(theta_estimation)):
    ax.plot(theta_difference(theta_estimation[i], theta, X), label=f'$\\tilde\\theta_{i+1}$')

ax.legend(loc='upper right', fontsize=12, ncol=2)

ax.set_title('Зависимость отклонения оценки параметра $|\\widetilde{\\theta}_{k} - \\theta|$ от количества элементов выборки $X_{n}$')

ax.set_xlabel('Объем выборки $n$', fontsize=12)
ax.set_ylabel('Отклонение оценки параметра $|\\widetilde{\\theta}_{k} - \\theta|$', fontsize=12)

fig.set_figheight(6)
fig.set_figwidth(12)

plt.show()
```
