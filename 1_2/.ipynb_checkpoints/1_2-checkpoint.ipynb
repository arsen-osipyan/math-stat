# Задание 2

Случайная величина имеет экспоненциальный закон распределения

$p(x) = \begin{cases} e^{-x}, x \ge 0,\\0, x \lt 0 \end{cases}$

Сгенерируйте выборку объема n = 25.

```python
import numpy as np
import random
from matplotlib import pyplot as plt
from matplotlib.patches import Rectangle
```

$x = ln(\frac{1}{1-F})$, где $F$ - случайная равномерно распределенная величина на $[0, 1)$.

```python
selection_size = 25

F = np.array([random.random() for _ in range(selection_size)])
X = np.array([np.log(1.0 / (1.0 - F[i])) for i in range(selection_size)])
X = np.array(list(map(lambda x: round(x, 3), X)))

print(X)
```

### a) Определить моду, медиану, размах, коэффициент асимметрии.

```python
def moment(X, k):
    a = [X[i]**k for i in range(len(X))]
    return np.mean(a)
        
def central_moment(X, k):
    alpha_1 = moment(X, 1)
    a = [(X[i] - alpha_1)**k for i in range(len(X))]
    return np.mean(a)

def asymmetry(X):
    return central_moment(X, 3) / central_moment(X, 2)**(3/2)

def variety(X):
    return central_moment(X, 2)
    
def std(X):
    return np.sqrt(central_moment(X, 2))

def modes(X):
    m = []
    unique, counts = np.unique(X, return_counts=True)
    max_count = np.max(counts)
    for i in range(len(unique)):
        if counts[i] == max_count:
            m.append(unique[i])
    return m


print(f'Моды выборки: {modes(X)}')
print(f'Медиана выборки: {round(np.median(X), 3)}')
print(f'Среднее значение выборки: {round(np.mean(X), 3)}')
print(f'Максимальное значение выборки: {np.max(X)}')
print(f'Минимальное значение выборки: {np.min(X)}')
print(f'Размах выборки: {round(np.max(X) - np.min(X), 3)}')
print(f'Коэффициент асимметрии выборки: {round(asymmetry(X), 3)}')
```

### b) Построить эмпирическую функцию распределения, гистограмму и boxplot.

```python
def empirical_probability(X, x):
    F = 0.0
    for i in range(len(X)):
        if (X[i] < x):
            F += 1
    F /= len(X)
    return F
```

```python
fig, ax = plt.subplots()

x = np.arange(np.min(X), np.max(X), 0.001)
y = np.array(list(map(lambda t: empirical_probability(X, t), x)))

ax.scatter(x, y, marker='x', s=2)
ax.set_title('Эмпирическая функция рапределения вероятности $\\tilde F(x)$')
ax.set_xlabel('x', fontsize=12)
ax.set_ylabel('$\\tilde F(x)$', fontsize=12)
fig.set_figheight(6)
fig.set_figwidth(12)
plt.show()
```

```python
def histogram(X):
    bars = []
    k = 1 + int(1 + np.log2(len(X)))
    borders = np.linspace(np.min(X), np.max(X), k+1)
    bars = np.histogram(X, bins=borders)
    bars = (np.array([s / (borders[1] - borders[0]) for s in bars[0]]), bars[1])
    return bars


fig, ax = plt.subplots()

bars = histogram(X)

ax.bar([f'[{round(bars[1][i], 3)}, {round(bars[1][i+1], 3)})' for i in range(len(bars[1])-1)], bars[0])
ax.set_title('Гистограмма для выборки $X_n$')
ax.set_xlabel('x', fontsize=12)
fig.set_figheight(6)
fig.set_figwidth(12)
plt.show()
```

```python
x25 = np.quantile(X, 0.25)
x75 = np.quantile(X, 0.75)
eps = x75 - x25
whis = [0,100]

if x25 - eps * 1.5 > np.min(X):
    a = 100
    for x in X:
        if x >= x25 - eps * 1.5:
            a -= 100 / len(X)
    whis[0] = a
if x75 + eps * 1.5 < np.max(X):
    a = 0
    for x in X:
        if x <= x75 + eps * 1.5:
            a += 100 / len(X)
    whis[1] = a

    
fig, ax = plt.subplots()
    
ax.boxplot(X, vert=0, whis=whis)

ax.set_title('Boxplot для выборки $X_n$')
ax.set_xlabel('x', fontsize=12)
fig.set_figheight(6)
fig.set_figwidth(12)
plt.show()
```

### c) Найти ядерную оценку плотности распределения.

```python
def nuclear_estimation(X):
    n = len(X)
    q = lambda y : 0.75*(1-y**2) if -1.0 <= y <= 1.0 else 0.0
    s = np.sqrt((n-1)/n*central_moment(X, 2))
    h = 2.344 * s / n**(1/5)
    x = np.arange(np.min(X), np.max(X), 0.001)
    y = [1/(n*h)*np.sum([q((x[i] - X[j])/h) for j in range(n)]) for i in range(len(x))]
    return x, y


fig, ax = plt.subplots()

x, y = nuclear_estimation(X)

ax.scatter(x, y, s=2)
ax.set_title('Ядерная оценка плотности распределения $\\tilde p (x)$ для выборки $X_n$')
ax.set_xlabel('x', fontsize=12)
ax.set_ylabel('$\\tilde p (x)$', fontsize=12)
fig.set_figheight(6)
fig.set_figwidth(12)
plt.show()
```

### d) Определить плотность распределения среднего арифметического элементов выборки. Сравнить с бутстраповской оценкой плотности.

Из ЦПТ Маркова легко вывести, что

$\frac{1}{n}\sum\limits_{i=1}^n x_{i} \leadsto N(M\xi,\frac{D\xi}{n})$

```python
def bootstrap_distribution(X, f, n_subselections=1000):
    n = len(X)
    subselections = np.zeros(n * n_subselections).reshape((n_subselections, n))
    fs = np.zeros(n_subselections)
    
    for i in range(n_subselections):
        for j in range(n):
            subselections[i][j] = X[random.randint(0, n-1)]
        fs[i] = f(subselections[i])
    
    k = 1 + int(1 + np.log2(n_subselections)) 
    fmax = np.max(fs)
    fmin = np.min(fs)
    scope = fmax - fmin
    borders = (fmin, fmax)
    delta = scope / k
    
    x = np.zeros(k)
    nus = np.zeros(k)
    
    for i in range(k):
        cur_borders = (borders[0] + i * delta, borders[0] + (i + 1) * delta)
        x[i] = cur_borders[0]
        for j in range(n_subselections):
            if cur_borders[0] <= fs[j] < cur_borders[1]:
                nus[i] += 1.0 / (n_subselections * delta)
    
    return (x, nus)


def normal_distribution(x, mean, std2):
    return 1.0 / (np.sqrt(2*np.pi*std2)) * np.exp(-(x - mean)**2 / (2 * std2))


fig, ax = plt.subplots()


x1, y1 = bootstrap_distribution(X, np.mean)
x2 = np.arange(np.min(X), np.max(X), 0.001)
mean = np.mean(X)
std2 = central_moment(X, 2) / len(X)
y2 = [normal_distribution(x, mean, std2) for x in x2]


for k in range(len(x1)):
    ax.add_patch(Rectangle((x1[k], 0), x1[1] - x1[0], y1[k], facecolor="#d5e6f7"))
ax.scatter(x2, y2, s=4)
ax.set_title("Плотность (график) и бутстраповская плотность (гистограмма) распределения\nсреднего арифметического $\\tilde p(x)$ для выборки $X_n$")
ax.set_xlabel('x', fontsize=12)
ax.set_ylabel('$\\tilde p (x)$', fontsize=12)
fig.set_figheight(6)
fig.set_figwidth(12)
plt.show()
```

### e) Найти бутстраповскую оценку плотности распределения коэффициента асимметрии.

```python
fig, ax = plt.subplots()

x, y = bootstrap_distribution(X, asymmetry)

ax.bar([f'[{round(x[i], 3), round(x[i] + x[1] - x[0], 3)})' for i in range(len(x))], y)

ax.set_xticklabels([f'[{round(x[i], 3)}, {round(x[i] + x[1] - x[0], 3)})' for i in range(len(x))], rotation=30)
ax.set_title("Бутстраповская оценка плотности распределения коэффициента ассиметрии $\\tilde p_\gamma (x)$")
ax.set_xlabel('x', fontsize=12)
ax.set_ylabel('$\\tilde p_\gamma (x)$', fontsize=12)
fig.set_figheight(6)
fig.set_figwidth(12)
plt.show()
```
